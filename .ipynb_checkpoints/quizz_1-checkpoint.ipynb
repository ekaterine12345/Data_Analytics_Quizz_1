**დიაგრამები**

```python
import matplotlib.pyplot as plt
```

```python
age_counts = df['Age'].value_counts(normalize=True)

age_categories = age_counts.index.tolist() # ასაკობრივი კატეგორიები ინახება ლისტის სახით
percents = (age_counts.to_numpy() * 100).round(2) # მეასედებამდე დამრგვალებული შესაბამისი პროცენტულობა ამ ასაკობრივი კატეგორიის
```

```python
fig, ax = plt.subplots(figsize =(8, 4)) # განვსაზღვრეთ ფიგურის ზომები
plt.title(schema_df.loc['Age', 'question'][:-2], fontsize=20) # სათაურად დავაწერეთ კითხვა ასაკის შესახებ

ax.barh(age_categories, percents, color="darkblue") # ჰორიზონტალური მუქი ლურჯი ბარ ჩართის აგება

for bor in ['top', 'bottom', 'left', 'right']: # საზღვრების მოშლა
    ax.spines[bor].set_visible(False)

ax.xaxis.set_ticks_position('none') #  x და y  პატარა ხაზების მოშლა
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad = 8) # პედინგების დამატება
ax.yaxis.set_tick_params(pad = 7)

ax.invert_yaxis() # პროცენტულობის კლების მიხედვით დალაგება

for i in ax.patches: # იტერირება ხდება - თითოეულ ბარზე (ჰორიზონტალურ სვეტზე) წვდომა და თითოეული პლოტიშთვის შესაბამისი პროცენტობის წაწერა
    plt.text(i.get_width()+0.3, i.get_y()+0.5, str(round((i.get_width()), 2))+"%",
             fontsize = 9, fontweight ='bold',
             color ='black')
```

```python
AISel_count = df['AISelect'].value_counts(normalize=True)
AISelect_labels = AISel_count.index.tolist()
AISel_percents = (AISel_count.to_numpy() * 100).round(2)
# მონაცემთა წამოღება და დაანგარიშება, რათა გავიგოთ:
# თუ რამდენი პროცენტი იყენებს ხელოვნური ინტელექტის თულებს; რამდენი გეგმავს რომ გამოიყენებს და რამდენი აცხადებს უარს გამოყენებაზე.
# ამ 3 პახუსი პროცენტულობის მიხედვით ავაგებთ წრიულ დიაგრამას
```

```python
plt.title(schema_df.loc['AISelect', 'question'][:-2], fontsize=20) # სათაურად იმ კითხვის დაწერა, რომელზე გაცემულ პასუხებსაც განვიხილავთ
plt.pie(AISel_percents, labels=AISelect_labels, shadow=True, colors=['darkgreen', 'Blue', 'yellow']) # წრიული დიაგრამის აგება
plt.legend(title="AI Usage",  bbox_to_anchor=(1.3,0.25))  # მაჯვენა ქვედა კუთხეში ვქმნით ლეგენდას
```

```python
courses_list = ['Udemy', 'Coursera', 'Codecademy', 'Pluralsight', 'edX', 'Udacity', 'Skillsoft'] # ჩვენთვის სასურველი კურსების ამორჩევა
courses_percent = []
for each in courses_list:
    courses_percent.append((df['LearnCodeCoursesCert'].str.contains(each).value_counts(normalize=True)[True]*100).round(2)) # ამ კურსის შესაბამისი
    # პროცენტული მაჩვენებლის დაანგარიშება და ლისტში დამატება. პროცენტულ მაჩვენებელში აქ იგულისხმება მთლიანობასთან შედარებით და არა ერთმანეთთან,
    # რადგანაც ერთი რესპოდენტი შესაძლოა ერთზე მეტ ონლაინ კურსის პლატფორმას იყენებდეს
courses_percent = np.array(courses_percent)
```

```python
fig, ax = plt.subplots(figsize =(16, 6))
plt.title("Percentage of developers using online courses platformas", fontsize=20) # დავასათაურეთ 
ax.bar(courses_list, courses_percent, color="orange", edgecolor="grey", linewidth=2) # ბარ ჩართის აგება

for bor in ['top', 'bottom', 'left', 'right']: # 
    ax.spines[bor].set_visible(False) # საზღვრების მოშლა

ax.xaxis.set_ticks_position('none') # მხოლოდ X ღერძისთვის პატარა ხაზების მოშლა

for i in ax.patches: # იტერირება ხდება - თითოეულ ბარზე(ვერტიკალურ სვეტზე) წვდომა და მათ ზემოთ შესაბამისი პროცენტულობის დაწერა
    plt.text(i.get_x()+0.25, i.get_height()+0.5, str(round((i.get_height()), 2))+"%",
             fontsize = 9, fontweight ='bold',
             color ='black')
```
